```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/home/code/deepcodecorrection/scripts/data_perf.csv", sep=" ")

data_other = pd.read_csv("/home/code/deepcodecorrection/scripts/data_perf_productAE.csv", sep=" ")
```

```python
code_rate_filtering = 0.5

def plot_ber_vs_snr(data, data_other, code_rate_filtering):

    plt.figure(figsize=(20, 10))

    for value_group, df_grp in data.groupby(["code_rate", "N", "METHOD"]):

        if value_group[0] == code_rate_filtering:
            plt.plot(df_grp["SNR"], df_grp["PERF"], marker='o', label="R = " + str(value_group[0]) + " with "  + str(value_group[2]))

    for value_group, df_grp in data_other.groupby(["code_rate", "N", "METHOD"]):
        if value_group[0] == code_rate_filtering:
            plt.plot(df_grp["SNR"], df_grp["PERF"], marker='o', label="R = " + str(value_group[0]) + " with "  + str(value_group[2]))


    plt.yscale('log')
    plt.title('BER vs SNR')
    plt.xlabel('SNR' + r'$\rightarrow$')
    plt.ylabel('BER' + r'$\rightarrow$')

    # Showing grid and highlighting axis at x=0 and y=0
    plt.grid(True, which="both", ls="-", color='0.65')
    plt.legend()
    plt.show()

plot_ber_vs_snr(data, data_other, 0.5)
plot_ber_vs_snr(data, data_other, 0.33)
```
